```python
from EEG import EEG
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

eeg = EEG(file='AC p1d0,8ed.csv')
eeg.process(['FP1-A1'])
```

```python
rho = {k: max(v) - min(v) for k, v in eeg.rho['FP1-A1'].items()}

plt.plot(rho.keys(), rho.values())
plt.title('Chunk size to |ρ| relation on FP1-A1 sensor')
plt.xlabel('Chunk size')
plt.ylabel('|ρ|')
plt.show()
```
